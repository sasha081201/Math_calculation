```python
def dich(a, b, F1):
    e = 0.0001
    counter = 0
    results = pd.DataFrame(columns =['a', 'b', 'x', 'F(a)', 'F(b)', 'F(x)'])
    x = (a + b)/2
    while(abs(a - b) > e):
        res = {'a': a, 'b': b, 'x': x, 'F(a)': F1(a), 'F(b)': F1(b), 'F(x)': F1(x)}
        counter += 1
        x = (a + b)/2
        if(F1(a) * F1(x) > 0):
            a = x
        else:
            b = x
        
        results.loc[counter-1] = res
    display(results)
    
dich(-0.5, 0, F1)
dich(4, 6, F1)
dich(-2, -1, F2)
dich(1, 2, F2)
```

```python
def secant(a, b, x0, F1):
    e = 0.0001
    results_secant = pd.DataFrame(columns =['a', 'b', 'x', 'F(a)', 'F(b)', 'F(x)'])
    counter = 0
    while(abs(F1(x0)) > e):
        res_secant = {'a': a, 'b': b, 'x': x0, 'F(a)': F1(a), 'F(b)': F1(b), 'F(x)': F1(x0)}
        counter += 1
        x0 = a -((b - a)* F1(a)/(F1(b) - F1(a)))
        if(F1(a) * F1(x0) > 0):
            a = x0
        else:
            b = x0
        results_secant.loc[counter-1] = res_secant
    display(results_secant)

secant(-1, 0, -0.5, F1)
secant(4, 5, 4.5, F1)
secant(-2, -1, -1.5, F2)
secant(1, 2, 1.5, F2)
```
